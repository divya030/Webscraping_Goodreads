#### Define a function get_genre_lists to get the genre lists with the help of a genre page

```python
def get_Genre_page(Genre_url):
    response = requests.get(Genre_url)
    #Check successful response
    if response.status_code != 200:
        raise Exception('Failed to load page {}'.format(Genre_url))
    #parse using BeautifulSoup   
    Genre_doc = BeautifulSoup(response.text,'html.parser')
    return Genre_doc


def get_genre_url():
    
    page_content = get_Genre_page(URL)
    
    genre_30 = page_content.find_all('a', class_ = 'actionLinkLite')[5:]
    
    list_of_genre = []

    for i in genre_30:
        list_of_genre.append(i.text.split()[0])
        
    genre_url = []
    
    for i in genre_30:
        url = "https://www.goodreads.com/" + i['href']
        genre_url.append(url)


    genre_dict_lst = {'title' : list_of_genre , 'url' : genre_url }

    genre_page_1 = pd.DataFrame(genre_dict_lst)
    
    genre_page_1.to_csv('Genre_types.csv',index = None)
    
    return genre_page_1


def get_genre_list():
    
    genre_url = get_genre_url()
    
    title_ = []
    list_title = []
    list_url = []
    votes= []
    books = []
    
    base_url = 'https://www.goodreads.com/'

    for i in genre_url['url']:
        Genre_content = requests.get(i)
        Genre_page = BeautifulSoup(Genre_content.content,"html.parser")
        list_content = Genre_page.find_all('a',class_ = 'listTitle')
        list_rating = Genre_page.find_all('div',class_ = 'listFullDetails')
        
        for j in list_content:
            title = i.split('/')[-1]
            title_.append(title)
            list_title.append(j.text)
            list_url.append(base_url + j['href'])
            
        for z in list_rating:
            vote = z.text.split()
            vo_ra = " ".join(vote).split("— ")
            votes.append(vo_ra[0].split(" ")[0])
            books.append(vo_ra[1].split(" ")[0])
    
    book_dict  ={'genre' : title_,
    'genre_listTitle': list_title,
    'url': list_url,
    'voters' : votes,
    'Books' : books }
    
    df = pd.DataFrame(book_dict)
    return df
```
