#### Define a function get_genre_lists to get the genre lists with the help of a genre page

```python
def get_Genre_page(Genre_url):
    response = requests.get(Genre_url)
    #Check successful response
    if response.status_code != 200:
        raise Exception('Failed to load page {}'.format(Genre_url))
    #parse using BeautifulSoup   
    Genre_doc = BeautifulSoup(response.text,'html.parser')
    return Genre_doc


def get_genre_url():
    
    page_content = get_Genre_page(URL)
    
    genre_30 = page_content.find_all('a', class_ = 'actionLinkLite')[5:]
    
    list_of_genre = []

    for i in genre_30:
        list_of_genre.append(i.text.split()[0])
        
    genre_url = []
    
    for i in genre_30:
        url = "https://www.goodreads.com/" + i['href']
        genre_url.append(url)


    genre_dict_lst = {'title' : list_of_genre , 'url' : genre_url }

    genre_page_1 = pd.DataFrame(genre_dict_lst)
    
    genre_page_1.to_csv('Genre_types.csv',index = None)
    
    return genre_page_1


def get_genre_list():
    
    genre_url = get_genre_url()
    
    title_ = []
    list_title = []
    list_url = []
    votes= []
    books = []
    
    base_url = 'https://www.goodreads.com/'

    for i in genre_url['url']:
        Genre_content = requests.get(i)
        Genre_page = BeautifulSoup(Genre_content.content,"html.parser")
        list_content = Genre_page.find_all('a',class_ = 'listTitle')
        list_rating = Genre_page.find_all('div',class_ = 'listFullDetails')
        
        for j in list_content:
            title = i.split('/')[-1]
            title_.append(title)
            list_title.append(j.text)
            list_url.append(base_url + j['href'])
            
        for z in list_rating:
            vote = z.text.split()
            vo_ra = " ".join(vote).split("— ")
            books.append(vo_ra[0].split(" ")[0])
            votes.append(vo_ra[1].split(" ")[0])
    
    book_dict  ={'genre' : title_,
    'genre_listTitle': list_title,
    'url': list_url,
    'Books' : votes,
    'votes' : books }
    
    df = pd.DataFrame(book_dict)
    return df
```

```python
fiction_url = genre_page_1['url'][1]

def get_Genre_page(Genre_url):
    response = requests.get(Genre_url)
    #Check successful response
    if response.status_code != 200:
        raise Exception('Failed to load page {}'.format(Genre_url))
    #parse using BeautifulSoup   
    Genre_doc = BeautifulSoup(response.text,'html.parser')
    return Genre_doc


def get_book_details():
    
    list_title = []
    list_url = []
    title = []
    book_url = []
    description = []
    author = []
    avg_rating = []
    total_rating = []
    score = []
    votes = []

    base_url = 'https://www.goodreads.com/'
    
    Genre_page = get_Genre_page(fiction_url)
    list_content = Genre_page.find_all('a',class_ = 'listTitle')
    
    for i in list_content:
        list_title.append(i.text)
        list_url.append(base_url + i['href'])
    
    Fiction_Best_series = get_Genre_page(list_url[6])
    Fiction_book_title  = Fiction_Best_series.find_all('a', class_ = 'bookTitle')
    Fiction_book_author  = Fiction_Best_series.find_all('a', class_ = 'authorName')
    Fiction_book_rating  = Fiction_Best_series.find_all('span', class_ = 'minirating')
    Fiction_book_score  = Fiction_Best_series.find_all('span', class_ = 'smallText uitext')

    for i in Fiction_book_title:
        title.append(i.text.strip())
        book_url.append(base_url + i['href'])
        
        desc = requests.get(base_url + i['href'])
        desc_page = BeautifulSoup(desc.content,"html.parser")
        des = desc_page.find('span',class_ = 'Formatted').text
        description.append(des)
        
    for i in Fiction_book_author:
        author.append(i.text)

    for i in Fiction_book_rating:
        avg_rating.append(i.text.split()[0])
        total_rating.append(i.text.split()[4])
    
    for i in Fiction_book_score:
        score.append(i.text.split()[1][:-1])
        votes.append(i.text.split()[3])
        
        
    Fiction_book_dict ={
    'Fiction_bookTitle': title,
    'Fiction_bookAuthor': author,
    'Book_url': book_url,
    'description': description,
    'AvgBookRating' :avg_rating,
    'TotalRatings': total_rating,
    'score': score,
    'votes': votes }

    fiction_df = pd.DataFrame(Fiction_book_dict)
    return fiction_df
   
```
